```python
import rasterio
from rasterio.warp import reproject, Resampling

# 1) Read NIR (10 m) with the same window as R,G,B
with rasterio.open(b08) as nir_src:
    nir_sub = nir_src.read(1, window=window).astype("float32")

print("nir_sub shape:", nir_sub.shape)

# 2) Read full SWIR (20 m) and reproject to the 10 m cropped grid
with rasterio.open(b11) as swir_src:
    swir_full = swir_src.read(1).astype("float32")

    # Create empty array to hold SWIR resampled to cropped R/G/B grid
    swir_sub = np.empty_like(red_sub, dtype="float32")

    reproject(
        source=swir_full,
        destination=swir_sub,
        src_transform=swir_src.transform,
        src_crs=swir_src.crs,
        dst_transform=sub_transform,  # transform of the cropped 10m window
        dst_crs=sub_crs,
        resampling=Resampling.bilinear,
    )

print("swir_sub shape:", swir_sub.shape)
```

```python
from pathlib import Path

# Stack 5 bands: [R, G, B, NIR, SWIR], scaled to [0,1]
multiband_sub = np.stack(
    [
        red_sub / 10000.0,
        green_sub / 10000.0,
        blue_sub / 10000.0,
        nir_sub / 10000.0,
        swir_sub / 10000.0,
    ],
    axis=-1,  # (H, W, 5)
)

multiband_sub = np.clip(multiband_sub, 0, 1)

print("multiband_sub shape:", multiband_sub.shape)

processed_dir = project_root / "data" / "processed"
processed_dir.mkdir(parents=True, exist_ok=True)

out_path = processed_dir / "gangotri_multiband_20181029.npy"
np.save(out_path, multiband_sub.astype("float32"))

print("Saved multiband array to:", out_path)
```
