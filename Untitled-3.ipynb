```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

project_root = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
processed_dir = project_root / "data" / "processed"

mb = np.load(processed_dir / "gangotri_multiband_20181029.npy")  # (H, W, 5)
mask = np.load(processed_dir / "gangotri_mask_20181029.npy")     # (H, W)

print("Multiband shape:", mb.shape)
print("Mask shape:", mask.shape, "glacier pixels:", mask.sum())
```

```python
import numpy as np

H, W, C = mb.shape
patch_size = 128
num_patches = 1000

X_patches = []
y_patches = []

rng = np.random.default_rng(42)

max_tries = num_patches * 10  # avoid infinite loops
tries = 0

while len(X_patches) < num_patches and tries < max_tries:
    tries += 1

    # Sample random center somewhere inside the image
    cy = rng.integers(patch_size // 2, H - patch_size // 2)
    cx = rng.integers(patch_size // 2, W - patch_size // 2)

    y0 = cy - patch_size // 2
    y1 = y0 + patch_size
    x0 = cx - patch_size // 2
    x1 = x0 + patch_size

    mb_patch = mb[y0:y1, x0:x1, :]      # (128, 128, 5)
    mask_patch = mask[y0:y1, x0:x1]     # (128, 128)

    # Require at least some glacier in the patch (e.g. >1% of pixels)
    glacier_frac = mask_patch.mean()  # since mask is 0/1
    if glacier_frac < 0.01:
        continue

    X_patches.append(mb_patch)
    y_patches.append(mask_patch)

print("Collected patches:", len(X_patches), "tries:", tries)
```

```python
import numpy as np

# Convert list -> array
X_patches = np.stack(X_patches, axis=0)   # (N, 128, 128, 5)
y_patches = np.stack(y_patches, axis=0)   # (N, 128, 128)

print("X_patches:", X_patches.shape, X_patches.dtype)
print("y_patches:", y_patches.shape, y_patches.dtype)

# Move channels to first dimension for PyTorch: (N, 5, 128, 128)
X_patches = np.transpose(X_patches, (0, 3, 1, 2))
# Add channel dimension to mask: (N, 1, 128, 128)
y_patches = y_patches[:, None, :, :]

print("X_patches (channels first):", X_patches.shape)
print("y_patches (with channel):", y_patches.shape)
```

```python
idx = np.random.randint(0, len(X_patches))

# Take first 3 channels (RGB) and convert back to HWC for display
rgb_patch = np.transpose(X_patches[idx, :3], (1, 2, 0))  # (128, 128, 3)
mask_patch = y_patches[idx, 0]                           # (128, 128)

fig, ax = plt.subplots(1, 2, figsize=(6, 3))
ax[0].imshow(rgb_patch)
ax[0].set_title("RGB patch")
ax[0].axis("off")

ax[1].imshow(mask_patch, cmap="gray")
ax[1].set_title("Mask patch")
ax[1].axis("off")

plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_patches, y_patches, test_size=0.2, random_state=42
)

print("Train:", X_train.shape, y_train.shape)
print("Val:",   X_val.shape,   y_val.shape)

processed_dir = project_root / "data" / "processed"

np.save(processed_dir / "X_train_patches.npy", X_train.astype("float32"))
np.save(processed_dir / "y_train_patches.npy", y_train.astype("uint8"))
np.save(processed_dir / "X_val_patches.npy",   X_val.astype("float32"))
np.save(processed_dir / "y_val_patches.npy",   y_val.astype("uint8"))

print("Saved patch datasets to:", processed_dir)
```
